imagem.ravel() converte a matriz para vetor

```python
image_features = []
image_class = []
for fruit_name in list_of_fruits_directory:
    dir_fruit = directory+fruit_name
    list_of_images = [dir_fruit +'/'+ image for image in os.listdir(dir_fruit)]
    for fruit_image in list_of_images:
        image_r = cv2.imread(fruit_image)
        image_r = image_r.ravel()
        image_features.append(image_r)
        image_class.append(fruit_name)
```

```python
x_image =  np.asarray(image_features)
y_image =  np.asarray(image_class)
```
